##Neophodne biblioteke

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

##Ucitavanje skupa podataka

```python
data = pd.read_csv("Real estate.csv")
data.head()
```

##Cena kuce po jedinici povrsine u zavisnosti od datuma transakcije, starosti kuce, razdaljine do najblize MRT stanice, broja prodavnica, geografske sirine i duzine

```python
sns.scatterplot(data, x=data['transaction date'], y=data['house price of unit area'])
plt.show()
sns.scatterplot(data, x=data['house age'], y=data['house price of unit area'])
plt.show()
sns.scatterplot(data, x=data['distance to the nearest MRT station'], y=data['house price of unit area'])
plt.show()
sns.scatterplot(data, x=data['number of convenience stores'], y=data['house price of unit area'])
plt.show()
sns.scatterplot(data, x=data['latitude'], y=data['house price of unit area'])
plt.show()
sns.scatterplot(data, x=data['longitude'], y=data['house price of unit area'])
plt.show()
```

##Tumacenje grafika

```python
cor = data.corr()
sns.heatmap(cor)
```
