```python
import pandas as pd
import missingno as msno
data = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
display(data.head(30))
msno.matrix(data)
print(len(data))
```

```python
import numpy as np
data['Age'] = data['Age'].fillna(data['Age'].mean())
#data=data.dropna(subset=['Age'])
data['Has Cabin'] = data['Cabin'].isna().apply(lambda x: not x)
data=data.replace('female',0)
data=data.replace('male',1)
data.drop(columns=["Cabin", "SibSp", "Parch", "Ticket"], inplace=True)
display(data.head(30))
msno.matrix(data)
```

```python
data.corr()['Survived']
```

```python
data = data.drop(columns=['Age', 'PassengerId'])
data.head(10)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

model = LogisticRegression()

dataset = data[['Pclass', 'Fare', 'Has Cabin', 'Sex', 'Survived']]

train, test = train_test_split(dataset, test_size=0.2, random_state=42)

Xtrain = train[['Pclass', 'Fare', 'Has Cabin', 'Sex']]
ytrain = train['Survived']

model.fit(Xtrain, ytrain)

Xtest = test[['Pclass', 'Fare', 'Has Cabin', 'Sex']]

y_pred = model.predict(Xtest)
y_true = test[['Survived']]

from sklearn.metrics import classification_report
print(classification_report(y_true, y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler

model = KNeighborsClassifier(n_neighbors=31)

dataset = data[['Pclass', 'Fare', 'Has Cabin', 'Sex', 'Survived']]


train, test = train_test_split(dataset, test_size=0.2, random_state=42)

Xtrain = train[['Pclass', 'Fare', 'Has Cabin', 'Sex']]
ytrain = train['Survived']

model.fit(Xtrain, ytrain)

Xtest = test[['Pclass', 'Fare', 'Has Cabin', 'Sex']]

y_pred = model.predict(Xtest)
y_true = test[['Survived']]

from sklearn.metrics import classification_report
print(classification_report(y_true, y_pred))
```
